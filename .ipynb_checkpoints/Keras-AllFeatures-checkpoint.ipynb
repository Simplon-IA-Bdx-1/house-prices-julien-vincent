```python
import numpy as np 
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras import metrics
import seaborn as sns
import matplotlib.pyplot as plt
import keras
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import IsolationForest
from tensorflow import random as tf_random
from tensorflow.keras.optimizers import Adam

# ignore Warning
import warnings
warnings.filterwarnings("ignore") 
```

```python
train = pd.read_csv('files_csv/trainkeras.csv',index_col=0)
test = pd.read_csv('files_csv/testkeras.csv',index_col=0)
```

```python
df = pd.read_csv('files_csv/test.csv')
test_ID = df['Id']
test_ID
```

#### X_train , y_train , X_test

```python
target_column = 'SalePrice'
y_train_full = train[target_column].values
X_train_full = train.drop(target_column, axis=1).values
X_test = test.values
```

#### Seed

```python
from numpy import random as np_random
SEED = 42
np_random.seed(SEED)
```

#### Split

```python
X_train, X_val, y_train, y_val = train_test_split(X_train_full, y_train_full)
```

#### Scale
On scale les dummies ou non ? 

```python
scaler_X = StandardScaler()
scaler_X.fit(X_train)
```

```python
X_train_scaled = scaler_X.transform(X_train)
X_val_scaled = scaler_X.transform(X_val)
X_test_scaled = scaler_X.transform(X_test)
```

```python
scaler_Y_log = StandardScaler()
scaler_Y_log.fit(y_train.reshape(-1,1))
```

```python
y_train_log_scaled = scaler_Y_log.transform(y_train.reshape(-1,1))
y_val_log_scaled = scaler_Y_log.transform(y_val.reshape(-1,1))
```

#### Model

```python
tf_random.set_seed(SEED)
```

```python
n_input = X_train.shape[1]
def create_model():
    # create model
    model = Sequential()
    model.add(Dense(220,input_dim=n_input, kernel_initializer='normal', activation='relu'))
    model.add(Dense(110, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))
    # Compile model
#     model.compile(loss='mean_squared_error', optimizer=keras.optimizers.Adadelta(learning_rate=0.005, rho=0.95))
    model.compile(loss='mean_squared_error', optimizer=keras.optimizers.Adam(learning_rate=0.01))
    return model
```

```python
model = create_model()
model.summary()
```

```python
history = model.fit(X_train_scaled, y_train_log_scaled, validation_data=(X_val_scaled,y_val_log_scaled), epochs=500, batch_size=20)
```

```python
pd.DataFrame(history.history).plot()
plt.grid(True)
plt.title('Model performance throughout training')
plt.ylabel('Loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper right')
plt.show()
```

#### Prédiction

```python
prediction_scaled = model.predict(X_test_scaled)
```

```python
log_prediction = scaler_Y_log.inverse_transform(prediction_scaled)
prediction = np.expm1(log_prediction)
prediction 
```

```python
submission = pd.DataFrame()
submission['Id'] = test_ID
submission['SalePrice'] =  prediction
submission.to_csv('Keras-Allfeatures.csv', index=False)
```
